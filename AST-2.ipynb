```python
q1 = data['living area'].quantile(0.25) #Q1
q3 = data['living area'].quantile(0.75) #Q3
```

```python
IQR = q3-q1
```

```python
upper_limit = q3+1.5*IQR
lower_limit =q1-1.5*IQR
```

```python
data.median()
```

```python
data['living area'] = np.where(data['living area']>upper_limit,30,data['living area'])
```
